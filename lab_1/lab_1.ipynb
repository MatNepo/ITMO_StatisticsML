## 1.1. Вероятностное пространство, формула Байеса

### Задание 1

Определить (с обоснованием), зависимы или независимы следующие события

#### (a) Несовместные события

События несовместимы если $A \cap B = \varnothing$

Для независимости событий необходимо, чтобы $P(A \cap B) = P(A)P(B)$

Из условия $P(A \cap B) = 0$, а $P(A)P(B)>0 \Rightarrow$ Если несовместные события имеют ненулевые вероятности — они всегда зависимы.

#### (b) События, образующие $\sigma$-алгебру $\Sigma$ в пространстве $(\Omega,\Sigma,P)$

В вероятностном пространстве $(\Omega,\Sigma,P)$ $\sigma$-алгебра $\Sigma$ - это семейство подмножеств множества, для которых определена  в-ть P.

Она удовлетворяет 3 св-вам:
1) $\Omega \in \Sigma$
2) Если $A \in \Sigma$, то $A^C \in \Sigma$
3) Если $A_1, A_2, ... \in \Sigma \Rightarrow \bigcup_{n=1}^{\infty} A_n \in \Sigma$

<details>
<summary>Пояснение</summary>

1) омега принадлежит сигма-алгебре
2) если событие A содержится в сигма-алгебре, то его дополнение тоже содержится в сигма-алгебре
3) если события $A_1, A_2, ...$ лежат в сигма-алгебре, то их счётное объединение лежит в сигма-алгебре

**Опр.** Пусть даны множества:

$$
A_1, A_2, A_3, ...
$$

то есть столько множеств, что их можно перенумеровать натуральными числами (их **счётно много**).

**Счётное объединение** этих множеств - это множество всех элементов, которые входят хотя бы в одно из них:

$$
\bigcup_{n=1}^{\infty} A_n
$$

</details>

<details>
<summary>Пример</summary>

Пример зависимых и независимых событий в пространстве $(\Omega, \Sigma, P)$

Рассмотрим вероятностное пространство $(\Omega, \Sigma, P)$.

Пусть $\Omega = \{1, 2, 3, 4\}$ 
и определим $\Sigma$ как минимальную $\sigma$-алгебру, содержащую событие $A$:

$\Sigma = \{\emptyset, \Omega, A, A^c\},$

где $A = \{1,2\}, \qquad A^c = \{3,4\}.$

**1. Пример зависимых событий**

Пусть вероятность равномерная:

$P(\{i\}) = \tfrac{1}{4}, \quad i=1,2,3,4$

Тогда

$P(A) = \tfrac{1}{2}, \qquad P(A^c)=\tfrac{1}{2}$

Но
$$
A \cap A^c = \emptyset,
\qquad
P(A \cap A^c) = 0.
$$

Проверяем независимость: $P(A)P(A^c) = \tfrac{1}{2}\cdot \tfrac{1}{2} = \tfrac{1}{4}.$

Сравниваем:

$P(A\cap A^c) = 0 \neq \tfrac{1}{4} = P(A)P(A^c)$

*Вывод:* события $A$ и $A^c$ *зависимы*.

**2. Пример независимых событий (с той же $\Sigma$!)**

Теперь возьмём другую вероятность $P$:

$P(A)=1, \qquad P(A^c)=0$

Проверяем независимость $A$ и $A^c$.

1) Левое выражение:
$A\cap A^c = \emptyset, \qquad P(A\cap A^c)=0$

2) Правое выражение:
$P(A)P(A^c)=1\cdot 0 = 0$

Равенство выполняется.

Аналогично:

$P(A\cap A) = P(A) = 1 = 1\cdot1 = P(A)P(A),$

$P(A^c\cap A^c) = 0 = 0\cdot 0 = P(A^c)P(A^c)$

*Вывод:* все события в $\Sigma$ являются независимыми при таком выборе $P$.

</details>

*Ответ*: События, образующие $\sigma$-алгебру $\Sigma$ могут быть как зависимыми, так и независимыми. $\sigma$-алгебра не определяет независимость, её определяет вероятность $P$

#### (c) События, имеющие одинаковую вероятность

Если события равновероятны, то $P(A)=P(B)$ и это никак не связано с независимостью.

*Пример*: события могут иметь одинаковую вероятность, но быть зависимыми (бросаем монету $A = {герб}$, $B = {решка}$).

### Задание 4

Два парохода должны подойти к одному и тому же причалу независимо друг от друга и равновозможно в течение суток. Определить вероятность того, что одному из них придется ожидать освобождения причала, если время стоянки первого парохода — 1 час, а второго — 2 часа.

Два парохода подходят к причалу независимо и равномерно:  
$$T_1,\,T_2 \sim U[0,24].$$

Первый занимает причал $1$ час, второй — $2$ часа.  
Перекрытие интервалов стоянки:

$$[T_1,\,T_1+1] \cap [T_2,\,T_2+2] \ne \varnothing.$$

Это верно, если:

$$T_2 < T_1 + 1,\qquad T_1 < T_2 + 2.$$

$\Rightarrow$

$$T_2 - T_1 < 1,$$
$$T_2 - T_1 > -2.$$

$\Rightarrow$ условие ожидания:

$$-2 < T_2 - T_1 < 1$$

это полоса шириной $3$ внутри квадрата $[0,24]^2$.

Площадь квадрата $S = 24^2 = 576.$

Область вне полосы (где ожидания нет) — два треугольника со стороной: $24 - 3 = 21.$

Их суммарная площадь: $2 \cdot \frac{21^2}{2} = 441.$

Площадь области ожидания: $576 - 441 = 135.$

Искомая вероятность:

$$P = \frac{135}{576} = \frac{15}{64}.$$

### Задание 5

В самолёт попадает $m$ снарядов. Каждый снаряд независимо попадает в одну из трёх частей самолёта:

- в первую часть — с вероятностью $p_1$,
- во вторую часть — с вероятностью $p_2$,
- в третью часть — с вероятностью $p_3$,

где $p_1 + p_2 + p_3 = 1$.

Самолёт считается поражённым, если выполнено хотя бы одно из условий:
- в первую часть попал хотя бы один снаряд (a);
- во вторую часть попало не менее двух снарядов (b);
- в третью часть попало не менее трёх снарядов (c).

Требуется найти условную вероятность поражения самолёта $P(A \mid m)$ при $m = 1,2,3,4$.

$P(A \mid m) = \dfrac{P(A \cap m)}{P(m)}$

Рассмотрим случаи:

$m = 1$

Поражение возможно только если единственный снаряд попал в (a).
$P(A \mid 1) = p_1$

$m = 2$

Поражение происходит в следующих случаях:
- 2 попадания в (b): вероятность $p_2^2$
- 1 попадание в (b) и 1 в (a): таких комбинаций $2$, вероятность $2p_2 p_1$
- 1 попадание в (c) и 1 в (a): тоже $2$ комбинации, вероятность $2p_3 p_1$
- 2 попадания в (a): вероятность $p_1^2$

$\Rightarrow P(A \mid 2) = p_2^2 + 2p_2 p_1 + 2p_3 p_1 + p_1^2$.

$m = 3$

Самолёт *не* поражён только если:  
- 0 попаданий в (a);  
- 1 попадание в (b);  
- 2 попадания в (c).  
Комбинаций $C_3^1 = 3$, вероятность непоражения: $3 p_2 p_3^2$.

$\Rightarrow P(A \mid 3) = 1 - 3 p_2 p_3^2$.

$m = 4$

При четырёх попаданиях невозможно избежать выполнения одного из условий поражения.

$\Rightarrow P(A \mid 4) = 1$.

**Ответ**:

$P(A \mid 1) = p_1$  
$P(A \mid 2) = p_2^2 + 2p_2 p_1 + 2p_3 p_1 + p_1^2$  
$P(A \mid 3) = 1 - 3 p_2 p_3^2$  
$P(A \mid 4) = 1$

## 1.2. Случайный вектор и числовые характеристики



### Задание 1

Пусть  
$$
f_{\xi}(x, y) = \frac{e^{-2|y|}}{\pi (1 + x^2)}
$$  
Является ли данная функция плотностью распределения случайного вектора?

Функция является плотностью тогда и только тогда, когда выполнено условие нормировки:
$$\displaystyle \iint_{\mathbb{R}^2} f_{\xi}(x,y)\,dx\,dy = 1$$
и она неотрицательна.

Нормировка означает, что суммарная вероятность на всей плоскости равна 1.

$\displaystyle f_{\xi}(x,y)=\underbrace{\frac{1}{\pi(1+x^2)}}_{\text{плотность Коши}} \cdot \underbrace{e^{-2|y|}}_{\text{распределение Лапласса, параметр } 2}.$

Проверяем нормировку каждой части:

$\displaystyle \int_{-\infty}^{\infty} \frac{1}{\pi(1+x^2)}\,dx = 1$
(плотность Коши уже нормирована).

$\displaystyle \int_{-\infty}^{\infty} e^{-2|y|}\,dy = 1$
(плотность Лапласса с параметром 2).

Тогда двойной интеграл равен произведению нормировок:
$\displaystyle \iint f_{\xi}(x,y)\,dx\,dy = 1 \cdot 1 = 1.$

Функция неотрицательна, поэтому $f_{\xi}(x,y)$ является корректной плотностью.

<summary>Формулы</summary>

$E(\eta^2)$ — это математическое ожидание квадрата случайной величины $\eta$.  
Так же как $E(\eta)$ — это среднее значение самой $\eta$,  
$E(\eta^2)$ — это среднее значение величины $\eta^2$.

Для дискретной случайной величины математическое ожидание вычисляется по формуле:  
$\displaystyle E(X)=\sum_x x\,P(X=x)$.

Аналогично:  
$\displaystyle E(X^2)=\sum_x x^2\,P(X=x)$.

В задаче для $\eta$ имеем три возможных значения: $-1,\,0,\,1$, поэтому:

$E(\eta^2)=(-1)^2\cdot P(\eta=-1) + 0^2\cdot P(\eta=0) + 1^2\cdot P(\eta=1)$.

Это используется для нахождения дисперсии:  
$\displaystyle D(X)=E(X^2)-(E(X))^2$.

Список формул, использованных в решении:

1. Математическое ожидание дискретной величины:  
   $\displaystyle E(X)=\sum_x x\,P(X=x)$.

2. Математическое ожидание произведения:  
   $\displaystyle E(XY)=\sum_{x,y} xy\,P(X=x,Y=y)$.

3. Дисперсия:  
   $\displaystyle D(X)=E(X^2)-(E(X))^2$.

4. Ковариация:  
   $\displaystyle \mathrm{Cov}(X,Y)=E(XY)-E(X)E(Y)$.

5. Корреляция:  
   $\displaystyle \rho_{X,Y}=\dfrac{\mathrm{Cov}(X,Y)}{\sqrt{D(X)D(Y)}}$.

6. Матрицы:  
   - ковариационная: $\displaystyle \Sigma=\begin{pmatrix}D(X)&\mathrm{Cov}(X,Y)\\[4pt]\mathrm{Cov}(X,Y)&D(Y)\end{pmatrix}$  
   - корреляционная: $\displaystyle R=\begin{pmatrix}1&\rho_{XY}\\[4pt]\rho_{XY}&1\end{pmatrix}$.

То есть $E(\eta^2)$ — это просто «берём квадрат каждого возможного значения, умножаем на его вероятность и суммируем», формула полностью стандартная.
</details>

#### (c) Независимость

Независимость требует:  
$P(\xi,\eta)=P(\xi)P(\eta)$ для всех значений.

Проверим, например, точку $(\xi=-1,\eta=-1)$:

$P(\xi=-1,\eta=-1)=\dfrac18.$  
$P(\xi=-1)P(\eta=-1)=\dfrac12\cdot\dfrac{11}{24}=\dfrac{11}{48}.$  

Так как $\dfrac18 \ne \dfrac{11}{48}$, величины зависимы.

Некоррелированность означает $Cov(\xi,\eta)=0$, но  
$Cov(\xi,\eta)=-\dfrac12\neq 0$,  
поэтому величины не являются некоррелированными.

Вывод: $\xi$ и $\eta$ зависимы и коррелированы.

$E(\varphi_1 \varphi_2)=\dfrac{3\cdot2 + 4\cdot2 + 5\cdot4 + 6\cdot2 + 7\cdot2}{16}

#### (d) Независимость и некоррелированность $\varphi_1$ и $\varphi_2$

### Задание 4

Пусть $\xi \sim U_{-\pi,\pi}$ и $\eta_1 = \cos \xi$, $\eta_2 = \sin \xi$.

(a) Вычислить математическое ожидание, ковариационную и корреляционную матрицы вектора $(\xi, \eta)$

(b) Исследовать $\xi$ и $\eta$ на независимость и некоррелированность

#### (a) Вычислить математическое ожидание, ковариационную и корреляционную матрицы вектора $(\xi, \eta)$

**Математические ожидания**

Так как $\xi \sim U[-\pi,\pi]$ — распределение симметрично, а $\cos\xi$ чётная, $\sin\xi$ нечётная:

$E\xi = 0.$  
$E(\cos\xi)=0.$  
$E(\sin\xi)=0.$

**Дисперсии**

$$
E(\xi^2)
= \frac{1}{2\pi}\int_{-\pi}^{\pi} x^2\,dx
= \frac{\pi^2}{3},
\qquad
D\xi=\frac{\pi^2}{3}.
$$

Для вычисления $E(\cos^2 \xi)$ используем тождество

$$
\cos^2 x = \frac{1 + \cos 2x}{2}.
$$

Так как $\xi \sim U[-\pi,\pi]$, плотность равна $f_\xi(x)=\frac{1}{2\pi}$.

$$
\Rightarrow
E(\cos^2\xi)
= \frac{1}{2\pi}\int_{-\pi}^{\pi} \cos^2 x \, dx
= \frac{1}{2\pi}\int_{-\pi}^{\pi} \frac{1+\cos 2x}{2}\,dx.
$$

Разделим интеграл:

$$
E(\cos^2\xi)
= \frac{1}{4\pi}\int_{-\pi}^{\pi} 1\,dx
+ \frac{1}{4\pi}\int_{-\pi}^{\pi} \cos 2x\,dx.
$$

Второй интеграл равен нулю, поэтому

$$
E(\cos^2\xi)
= \frac{1}{4\pi} \cdot 2\pi = \frac12.
$$

Поскольку $E(\cos\xi)=0$, то

$$
D(\eta_1)=\frac12.
$$

Аналогично для синуса:

$$
\sin^2 x = \frac{1 - \cos 2x}{2}.
$$

$$
E(\sin^2\xi)
= \frac{1}{2\pi}\int_{-\pi}^{\pi} \frac{1-\cos 2x}{2}\,dx
= \frac{1}{4\pi}\int_{-\pi}^{\pi} 1\,dx
- \frac{1}{4\pi}\int_{-\pi}^{\pi} \cos 2x\,dx.
$$

Второй интеграл равен нулю, значит

$$
E(\sin^2\xi)=\frac12.
$$

Так как $E(\sin\xi)=0$, то

$$
D(\eta_2)=\frac12.
$$

**Ковариации**

Функции под интегралами либо нечётные, либо интегралы обнуляются по симметрии:

$$
Cov(\xi,\eta_1)=0,\qquad Cov(\xi,\eta_2)=0,\qquad Cov(\eta_1,\eta_2)=0.
$$

**Ковариационная матрица**

$$
\Sigma =
\begin{pmatrix}
\frac{\pi^2}{3} & 0 & 0 \\
0 & \tfrac12 & 0 \\
0 & 0 & \tfrac12
\end{pmatrix}
$$

**Корреляционная матрица**

$$
R =
\begin{pmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{pmatrix}
$$

#### (b) Независимость и некоррелированность

$\eta_1$ и $\eta_2$ выражаются через $\xi$ по формулам  
$\eta_1=\cos\xi$, $\eta_2=\sin\xi$.

Между ними существует тождественная связь:

$$
\cos^2\xi + \sin^2\xi = 1.
$$

$\Rightarrow$

- величины **зависимы**,  
- но их ковариация равна нулю, поэтому они **некоррелированы**.

**Итог:** $\xi$, $\eta_1$, $\eta_2$ — зависимые, но попарно некоррелированные величины.

### Задание 5

Найти плотность распределения суммы двух независимых случайных величин $\xi$ и $\eta$, если $\xi \sim \mathrm{Exp}_2$ и $\eta \sim U_{0,1}$.

Пусть $\xi$ и $\eta$ независимы,  
$\xi \sim \mathrm{Exp}_2$ — экспоненциальное распределение с параметром $\lambda=2$,  
$\eta \sim U_{0,1}$ — равномерное распределение на $[0,1]$.

Найти плотность распределения суммы  
$$
Z = \xi + \eta.
$$

Плотность экспоненциальной случайной величины:
$$
f_\xi(x) = 2 e^{-2x},\quad x \ge 0.
$$

Плотность равномерной величины:
$$
f_\eta(y) = 1,\quad y \in [0,1].
$$

Так как величины независимы, используем формулу свёртки:
$$
f_Z(z) = \int_{-\infty}^{\infty} f_\xi(z-y)\, f_\eta(y)\,dy.
$$

Ограничения на область интегрирования:  
$f_\eta(y)=1$ только при $y\in[0,1]$,  
$f_\xi(z-y)>0$ только если $z-y\ge0$ ⇒ $y \le z$.

**Случай $0 \le z \le 1$**

Тогда $y \in [0,z]$:

$$
f_Z(z)=\int_0^z 2 e^{-2(z-y)}\,dy
= 2e^{-2z} \int_0^z e^{2y}\,dy.
$$

$$
\int_0^z e^{2y}\,dy = \frac12(e^{2z}-1),
$$

поэтому

$$
f_Z(z)=1 - e^{-2z}.
$$

**Случай $z \ge 1$**

Теперь $y \in [0,1]$:

$$
f_Z(z)=\int_0^1 2 e^{-2(z-y)}\,dy
= 2e^{-2z} \int_0^1 e^{2y}\,dy.
$$

$$
\int_0^1 e^{2y}\,dy = \frac12(e^2 - 1),
$$

следовательно,

$$
f_Z(z) = (e^2 - 1)\, e^{-2z}.
$$

**Случай $z < 0$**

Сумма $\xi+\eta$ не может быть отрицательной:

$$
f_Z(z)=0.
$$

**Итоговая плотность:**

$$
f_Z(z)=
\begin{cases}
0, & z < 0, \\[6pt]
1 - e^{-2z}, & 0 \le z \le 1, \\[6pt]
(e^2 - 1)\, e^{-2z}, & z \ge 1.
\end{cases}
$$
